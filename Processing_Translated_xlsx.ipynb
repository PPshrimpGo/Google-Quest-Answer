```python
def combine_row(file="train_body_german.xlsx", column="question_body"):
    
    data = pd.read_excel(file)
    
    data['new_index'] = data.index
    for i in range(data.index.shape[0]):
        if (isinstance(data.index[i], str)):
            print(data.index[i][:-1], data.index[i])
            data['new_index'][i] = float(data.index[i][:-1])
            
    data = data.set_index('new_index')
    data.index.names = ['index']
    
    print(data.index)
    
    data = data.dropna()
    print(data.shape)
    print(data.head())
    
    i = 0
    while (i < data.shape[0]):


        print("processing ", data.iloc[i].name, " of ", 6079)

        if (not np.isnan(float(data.iloc[i].name))):

            current_index = i
            content = data.iloc[i][column]

            i += 1

            if (i >= data.shape[0]):
                break
            
            while (np.isnan(float(data.iloc[i].name))):
                content += str(data.iloc[i][column])
                i += 1
                if (i >= data.shape[0]):
                    break

            data.iloc[current_index] = content
            
    data = data.reset_index().dropna().set_index('index')
    print(data.index)
    
    data.to_excel(file + "_clean.xlsx")
    
    return data
    
```

```python
data = combine_row(file="train_body_chinese.xlsx", column="question_body")
```
