# Combine rows after translation

```python
import pandas as pd
import numpy as np
```

```python
def combine_row(file="train_body_german.xlsx"):
    
    data = pd.read_excel(file)
    
    print(data.index)
    
    data = data.dropna()
    print(data.head())
    
    i = 0
    while (i < data.shape[0]):


        print("processing ", data.iloc[i].name, " of ", 6079)

        if (not np.isnan(data.iloc[i].name)):

            current_index = i
            content = data.iloc[i].question_body

            i += 1

            if (i >= data.shape[0]):
                break

            while (np.isnan(data.iloc[i].name)):
                content += str(data.iloc[i].question_body)
                i += 1

            data.iloc[current_index] = content
            
    data = data.reset_index().dropna().set_index('index')
    print(data.head())
    
    return data
    
```

```python
data = combine_row(file="train_body_german.xlsx")
```

```python
data.to_excel("train_body_german_clean.xlsx")
```
